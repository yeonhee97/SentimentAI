```python
import pandas as pd
import numpy as np
from urllib import parse

from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
import time
import datetime
from selenium import webdriver
```

```python
#드라마(코미디)

driver = webdriver.Chrome(r"C:\Users\rhkr8\OneDrive\Desktop\졸업작품\chromedriver_win32\chromedriver.exe")
driver.get("https://www.tving.com/vod/genre")
drama = driver.find_element_by_xpath('//*[@id="sub_filter_category2"]/div/ul/li[2]/a')
driver.execute_script("arguments[0].click();", drama)

drama2 = driver.find_element_by_xpath('//*[@id="sub_filter_category2_sub"]/div/ul/li[2]/a')
driver.execute_script("arguments[0].click();", drama2)

whileSeconds=20

start = datetime.datetime.now()
end = start + datetime.timedelta(seconds=whileSeconds)
while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(1)
    if datetime.datetime.now() > end:
        break

html = driver.page_source    
soup = BeautifulSoup(html, "html.parser")
```

```python
comedyhref=[]
comedytitle=[]
comedyimg=[]

base_url= "https://www.tving.com/"
list = soup.find_all("div", {"class": "program-item__info"})
list2 = soup.find_all("div",{"class":"program-item column-6 "})
for doc in list:
    ans = doc.find("a")
    href = doc.find("a").get('href')
    href = parse.urljoin(base_url,href)
    word = ans.text
    word = word.strip()
    comedytitle.append(word)
    comedyhref.append(href)
    
for doc in list2:
    img = doc.find('img')
    img2 = img['src']
    comedyimg.append(img2)
```

```python
comedysum=[]
for i in range(len(comedytitle)):
    comedysum.append((comedytitle[i],comedyimg[i],comedyhref[i]))
```

```python
comedysum[:5]
```

```python
df1 = pd.DataFrame(comedysum,columns=['영화 제목','영화 이미지','영화 링크'])
df1.to_csv('comedy_drama.csv',encoding='utf-8',index=False)
```
