# 드라마

```python
#드라마
dramatitle=[]
dramahref=[]
dramaimg=[]

for i in range(1, 5):
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20210116&tg=1&page="+str(i)
    html = urlopen(url)
    base_url = "https://movie.naver.com/"
    
    soup = BeautifulSoup(html.read(), "html.parser")
    
    docs = soup.find_all("div", {"class": "tit5"})
    
    for doc in docs:
        ans = doc.find("a")
        href = doc.find("a").get('href')
        href = parse.urljoin(base_url,href)
        word = ans.text
        word = word.strip()
        dramahref.append(href)
        dramatitle.append(word)
```

```python
for i in dramahref:
    url = i
    html = urlopen(url)
    
    soup = BeautifulSoup(html.read(), "html.parser")
    docs = soup.find_all("div", {"class": "poster"})
    for doc in docs:
        img=doc.find('img')
        img2 = img['src']
        dramaimg.append(img2)
```

```python
dramaimg2 = []
for i in range(len(dramaimg)):
    if(i%2 == 1):
        dramaimg2.append(dramaimg[i])
```

```python
dramasum=[]
for i in range(0,100):
    dramasum.append((dramatitle[i],dramaimg2[i],dramahref[i]))
```
