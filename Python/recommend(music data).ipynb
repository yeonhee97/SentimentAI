```python
#댄스
dtitle=[]
dsinger = []
dance=[]
dmusicimage=[]
dmlink = []
dmlink2 = []

for i in range(1,200,50):
    hdr = {'User-Agent' : 'Mozilla/5.0'}
    url = "https://www.melon.com/genre/song_list.htm?gnrCode=GN0200&steadyYn=Y#params%5BgnrCode%5D=GN0200&params%5BdtlGnrCode%5D=&params%5BorderBy%5D=NEW&params%5BsteadyYn%5D=Y&po=pageObj&startIndex="+str(i)

    req = urllib.request.Request(url,headers=hdr)
    html = urlopen(req)
    #base_url = "https://movie.naver.com/"

    soup = BeautifulSoup(html.read(), "html.parser")

    titles = soup.find_all('div', {"class": "ellipsis rank01"})
    singers = soup.find_all('div', {"class": "ellipsis rank02"})  
    imgurl = soup.find_all("div",{"class","wrap"})
    mslink = soup.find_all("a",{"class","image_typeAll"})
    
    for t in titles:
        ans = t.find('a')
        word = ans.text
        dtitle.append(word)
        
    for s in singers:
        ans2 = s.find('span', {"class": "checkEllipsis"})
        word2 = ans2.text
        dsinger.append(word2)
        
    for i in imgurl:
        im = i.find("img")
        if(im != None):
            dmusicimage.append(im['src'][:69])
            
    for i in mslink[3:]:
        if(i != None and i['href'][22:35] == "goAlbumDetail"):
            a = int(re.findall("\d+",i['href'][37:45])[0])
            dmlink.append(a)
            
    for i in dmlink:
        link = "https://www.melon.com/album/detail.htm?albumId="+str(i)
        dmlink2.append(link)
        
    
    for i in range(1,len(dtitle)):
        dance.append((dtitle[i],dsinger[i],dmusicimage[i],dmlink2[i]))
```

```python
len(dance)
```

```python
df2 = pd.DataFrame(dance,columns=['노래','가수','이미지','링크'])
df2.to_csv('dance.csv',encoding='utf-8',index=False)
```
